```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("healthcare_cost_sample.csv")

# Preview data
print("Shape:", df.shape)
df.head()
```

```python
# Check datatypes & nulls
df.info()

# Describe numerical columns
df.describe()

# Total cost per specialty
df.groupby('Specialty')['Total_Cost'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Total_Cost', y='Specialty', data=df.sort_values(by='Total_Cost', ascending=False))
plt.title("Total Cost by Specialty")
plt.xlabel("Total Cost ($)")
plt.ylabel("Specialty")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='State', y='Total_Cost', data=df)
plt.title("Cost Distribution by State")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='Total_Procedures', y='Total_Cost', hue='Specialty')
plt.title("Total Procedures vs Total Cost")
plt.show()
```
